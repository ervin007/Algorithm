```python
import pandas as pd
import numpy as np
import os
```

```python
# Files = ['1h_Intervals_AllPAtients.xlsx',]
#  '2h_Intervals_AllPAtients.xlsx',
#  '4h_Intervals_AllPAtients.xlsx',
#  '5h_Intervals_AllPAtients.xlsx',
#  '6h_Intervals_AllPAtients.xlsx',
#  '8h_Intervals_AllPAtients.xlsx']
Files = ["Results_153_Patients.xlsx"]
```

```python
Files.append("8h_Interval_Results/" + os.listdir("8h_Interval_Results/")[1])
```

```python
Files
```

```python
def Healthy(Patient_ID):
    global Patients_Diabetes_Type
    return Patients_Diabetes_Type[Patient_ID]["Healthy"]
def Glucose(Patient_ID):
    global Patients_Diabetes_Type
    return Patients_Diabetes_Type[Patient_ID]["HgA1C(%)"]
def Diabetes_Type_Patient(Patient_ID):
    global Patients_Diabetes_Type
    return Patients_Diabetes_Type[Patient_ID]["Diabetes_Type"]
```

```python
Datas = pd.read_excel(Files[0])
Datas.head()
```

```python
Datas["Healthy"] = Datas["ID"].apply(Healthy)
Datas["HgA1C(%)"] = Datas["ID"].apply(Glucose)
Datas["Diabetes_Type"] = Datas["ID"].apply(Diabetes_Type_Patient)
```

```python
Datas["End"].unique()
```

```python
def New_Class(X):
    try:
        if float(X) < 6.5:
            return "G"
        else:
            return "B"
    except:
        return "-"

# Datas.drop(Datas.loc[Datas["HgA1C(%)"] == "-"].index, inplace=True)

Datas["New_Class"] = Datas["HgA1C(%)"].apply(New_Class)
Datas.head(11)
```
